```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
image1 = cv2.imread('images/Q2/1.jpg' )
image2 = cv2.imread('images/Q2/2.jpg' )
image3 = cv2.imread('images/Q2/3.jpg' )
image4 = cv2.imread('images/Q2/4.jpg' )
image5 = cv2.imread('images/Q2/5.jpg' )
image6 = cv2.imread('images/Q2/6.jpg' )
image7 = cv2.imread('images/Q2/7.jpg' )
image8 = cv2.imread('images/Q2/8.jpg' )
i = 0
images = [image1 , image2 , image3 , image4 , image5 , image6 , image7 , image8]

fig, (ax1, ax2 , ax3,ax4,ax5,ax6,ax7 , ax8) = plt.subplots(nrows=1, ncols=8,figsize=(16, 16) ,
                                    sharex=True, sharey=True)
i = 0
for a in (ax1 , ax2, ax3, ax4, ax5, ax6 , ax7 , ax8):
        images[i]= cv2.resize(images[i] , (0,0) ,fx=0.4 , fy=0.4)
        a.imshow(images[i])
        a.set_axis_off()
        i +=1
   
plt.tight_layout()
plt.show()
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
stitcher = cv2.Stitcher_create(0)
st, stitch = stitcher.stitch(images)
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python
if st == 0:
   plt.imshow(stitch)
   cv2.imwrite('images/panaroma.png' , stitch)

else:
    print("stitching failed".format(st))
```
