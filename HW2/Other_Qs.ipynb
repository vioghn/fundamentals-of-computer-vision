```python
slice1 = np.uint8(img_blur)
edges = cv2.Canny(slice1,90,110)
plt.title('canny edge detector')
plt.imshow(edges)
plt.axis('off')
```

```python
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
```

```python
answer = np.arctan2(sobelx , sobely)*180 /np.pi
gradiantvalue =  cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
plt.imshow(gradiantvalue , cmap = 'gray')
plt.axis('off')

print('gradiant degree' , answer)
```

```python
def write_text(image , text , point) : 
    cv2.putText(image,text , point , cv2.FONT_HERSHEY_DUPLEX, 0.4, (0, 0, 0), 1)
```

```python
def find_vertices(image):
    # Your code goes here.
    im = cv2.Canny(image, 100, 140)
    ret, thresh = cv2.threshold(im, 127, 255, 0)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    biggest = None
    max_area = 0
    approx_contour = None
    triangle = None
    square = None
    circle = None
    point_x = 0
    point_y = 0

    
    list = [] 
    for i in contours:
        area = cv2.contourArea(i)
        perim = 0.01 * cv2.arcLength(i, True)
        x, y, w, h = cv2.boundingRect(i) 
        approxpoly = cv2.approxPolyDP(i,  perim, True)
        point_x = approxpoly[0][0][0]
        point_y = approxpoly[0][0][1]
        ratio = float(w)/h

        if len(approxpoly) == 3 : 
            if(triangle is None):
                write_text(image , 'triangle', (point_x, point_y))
            triangle = approxpoly; 

        if len(approxpoly) == 4 and area > max_area:
                if(approx_contour is None):
                    write_text(image , 'rectangle', (point_x, point_y))
                biggest = approxpoly
                max_area = area
                approx_contour = biggest
                
                

        if len(approxpoly) == 4 and ratio == 1:  
            if(square is None):
                    write_text(image , 'squere', (point_x, point_y))
            square = approxpoly
        

        
    
        if(len(approxpoly) >= 6):
             if(circle is None):
                     write_text(image , 'circle', (point_x, point_y))
             circle = approxpoly

        
            

    list = [approx_contour , triangle , square , circle  ]
        
                                                                                             
    return list
```

```python
# Let's draw the points on the original image.

list_vertices = find_vertices(im)

for vertices in list_vertices:
    for v in vertices:
                plt.scatter([x for x, y in v], [y for x, y in v])

plt.imshow(im , cmap = 'gray')
plt.axis('off')
```
